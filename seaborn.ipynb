### setup


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
%reload_ext autoreload
%autoreload
```

### see all the datasets 

```python
sns.get_dataset_names()
crash_df = sns.load_dataset("car_crashes")
crash_df
```

### DISTRIBUTION PLOT 



```python
sns.displot(crash_df['no_previous'],kde=True,bins=25)
```

```python
# a jointplot is related to distribution plots because it visualizes two variables' distributions and their relationship
sns.jointplot(x="alcohol",y="speeding",data=crash_df,kind="reg")
#kind  = kde,hex
```

```python
#NOW SOME KDE PLOT 
sns.kdeplot(crash_df['alcohol'])
#we can also make some kde plot for 2 variables
```

```python
# PAIR PLOT
sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')
tips_df

sns.pairplot(tips_df,hue="sex", palette="Blues")


```

```python
#RUG PLOT 
sns.rugplot(tips_df["tip"])
```

### styling 

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,4))
sns.jointplot(x="alcohol",y= "speeding",data=crash_df,kind="reg")
sns.set_context("paper")
#FOR PRESENTATION WE CAN CHANGE THE FONT SIZE TO MAKE THIS ONE BETTER FOR A POSTER
#to remove the spine
# sns.despine(left=True)
```

### CATEGORICAL DATA PLOTS

```python
sns.barplot(x="sex",y='total_bill',data=tips_df,estimator=np.std)
#estimator std and also np.median etc
```

```python
#COUNT PLT
sns.countplot(x="sex",data=tips_df)
#same like the bar plot but this one shows the count of the variables (estimator is on stand one set)
```

```python
#BOX PLOT
sns.boxplot(x="day",y='total_bill',data=tips_df,hue="sex")
```

```python
# VOILON PLOT COMBINATION OF BOX AND KDE PLOT (kde estimation)
sns.violinplot(x="day",y='total_bill',data=tips_df,hue="sex")
# we can split it with making it true
```

```python
# STRIP PLOT 
plt.figure(figsize=(8,5))
sns.stripplot(x="day",y='total_bill',data=tips_df,dodge=True, hue="sex")
# jitter to spread it out
```

```python
# SWARM PLOT 
sns.violinplot(x="day",y='total_bill',data=tips_df,)
sns.swarmplot(x="day",y='total_bill',data=tips_df,)
```

```python
# PALLET 
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x="day",y='total_bill',data=tips_df,hue="sex",palette="afmhot")
plt.legend(loc=0)
```

### MATRIX PLOT


```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(flights,cmap="Blues",linecolor="white",linewidths=1)
```

```python
iris= sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue="species")
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

```python
# FACET GRID
g = sns.FacetGrid(crash_df, col="speeding", col_wrap=3, height=4)
g.map(sns.histplot, "alcohol", kde=True)
```
