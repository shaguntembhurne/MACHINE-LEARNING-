### packages used

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/shaguntembhurne/MACHINE-LEARNING-/drawndata1.csv")
df.head()
```

```python
x = df[["x","y"]].values
y = df["z"]=="a"
```

```python
plt.scatter (x[:,0],x[:,1],c=y);
```

```python
from sklearn.preprocessing import StandardScaler
x_new = StandardScaler().fit_transform(x)
plt.scatter (x_new[:,0],x_new[:,1],c=y);
```

```python
from sklearn.preprocessing import quantile_transform
x_new = quantile_transform(x,n_quantiles=100)
plt.scatter (x_new[:,0],x_new[:,1],c=y)
```
