### importing library


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### CREATING A DUMMY WEATHER DATA 


```python
data = {
    "date" : pd.date_range(start="2024-11-01",periods=10),
    "Tempreture (C)": np.random.randint(15,30,10),
    "Humidity (%)": np.random.randint(30,90,10)
}
```

### Making the dataframe 

```python
df = pd.DataFrame(data)

mean_temp = np.mean(df["Tempreture (C)"])
mean_humidity = np.mean(df["Humidity (%)"])
```

### plotting the Tempreture trend


```python
plt.figure(figsize=(10,5))
plt.plot(df['date'],df["Tempreture (C)"],label = "Temp trend",color = "red",marker = "o")
plt.title('Tempreture Trend')
plt.grid(True)
plt.xlabel("Date")
plt.ylabel('Tempreture')
plt.legend()
plt.show()
```

### Bar graph for humidity analysis

```python
plt.figure(figsize=(20,10))
plt.bar(df['date'],df['Humidity (%)'])
plt.xlabel("DATE")
plt.ylabel('HUMIDITY')
plt.title('HUMIDITY ANALYSIS')
plt.show()
```
