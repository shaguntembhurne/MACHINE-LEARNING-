```python
import matplotlib.pyplot as plt
```

### LINE PLOT

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y,label = "line",color = 'blue', marker= 'o')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### BAR CHART


```python
categories = ["A","B","A",]
sales = ["5","7","9",]
plt.bar(categories,sales)
plt.title('SALES CHART')
plt.xlabel('sales')
plt.ylabel('categories')
plt.show()
```

### SCATTER PLOT

```python
x = [1,2,3,4,5]
y = [5,7,6,8,7]
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```

### PIE CHART


```python
labels = ["A","B","C"]
sizes = [20,30,40]
plt.pie(sizes, labels = labels, autopct = '%1.1f%%')
plt.title('PIE CHART')
plt.show()
```

```python
data  = [1,2,2,3,3,3,4,4,4,4,5]
plt.hist(data, bins=5)
plt.show()
```

### Customizations

```python
# colours and makers 
plt.plot(x,y,color="red",marker='o',linestyle='--')
#grid and lenegend
plt.grid(True)
#IMPORTING LEGEND
plt.legend(['DATASET1'])

```

```python
#SUBPLOT
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
categories = ['A', 'B', 'C', 'D']
sales = [15, 10, 25, 30]

# Create subplots
fig, axs = plt.subplots(2, 1)  # 2 rows, 1 column

# Line plot in the first subplot
axs[0].plot(x, y, marker='o')
axs[0].set_title("Line Plot")  # Set title for the first subplot

# Bar plot in the second subplot
axs[1].bar(categories, sales)
axs[1].set_title("Bar Chart")  # Set title for the second subplot

# Show the plots
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
